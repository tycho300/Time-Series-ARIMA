## Time Series Analysis with ARIMA

### Research Question 
Using the revenue data from the telecom dataset and time series analysis, can we predict revenue for 30 days past the two year data set?

The data set provided includes two years of daily revenue for the telcom company. The goals of this project are to use time time series analysis to both examine the time series of the revenue data for trends and seasonality, build a model based on the characteristics of the time series, and to use the model predict the next 30 days of revenue for the company. 

### Assumptions of Time Series Analysis

<u> Stationarity </u> 

The primary assumption for Time Series Analysis is that the time series is stationary. A stationary time series has a constant mean and variance, contains no trends, and has a constant pattern of autocorrelation. Autocorrelation refers to the correlation between the time series and a lagged version of itself. Having a constant pattern of autocorrelation means that the autocorrelation function at any lag is in the confidence interval.

### C3. Evaluation of Stationarity 

Based on the plot above it is clear that the time series is not stationary as there appears to be an overall upward trend to the revenue data over time. The non-stationarity of the time series can be confirmed with the autocorrelation function seen below. The autocorrelation for up to 20 lags shows autocorrelation outside the .05 confidence for every lag which indicates that the time series is not stationary. The second method used to confirm non-stationarity was the Dickey Fuller statistical test. This test identifies if a series has a unit root indicating that its variance changes over time. The null hypothesis of the test is that there is a unit root present in the series and is non-stationary. In order to reject the null hypothesis the p-value must be less than .05. In the test below, the p-value for the DF test is .07. This indicates that the null hypothesis should be accepted and the series is non-stationary. Finally, a visualization of the rolling mean and standard deviation shows that neither is constant, which indicates the time series is non-stationary. (See visualizations below.)



```python
# Plot autocorrelation functin (D'Amico, 2023)

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

plot_acf(revenue, lags=20)
plt.title("Autocorrelation Function (ACF)")
plt.xlabel("Lags")
plt.ylabel("ACF Value")
```

```python
# Run Dicky Fuller Test 
dftest = adfuller(revenue, autolag='AIC', regression='ct')

# Print the P-Value for the result
print("The P-Value for the DF test is: ", dftest[1])

# Create conditions for p-value to reject or accept the null hypothesis. 
if dftest[1] < .05:
    print("The p value is less than .05, reject null hypothesis. Series is stationary.")
else :
    print("The p value is greater than .05, accept null hypothesis. Series is non-stationary.")
```

```python
# Plot the rolling mean to identify trend 
import seaborn as sns
import warnings
warnings.filterwarnings(action="ignore")

# Create copy of dataset to add the rolling average. 
rolling_stats = revenue.copy()

# Add columns for rolling mean and rolling standard deviation at 12 day intervals. 
rolling_stats['rmean'] = rolling_stats.rolling(12).mean()
rolling_stats['rstd'] = rolling_stats['Revenue'].rolling(12).std()

# Create plot to compare rolling average with the time series data
plt.figure(figsize = [12 , 6])
plt.title("Rolling Statistics for Time Series")
sns.lineplot(x='Day', y='Revenue', data=rolling_stats, label='Series')
sns.lineplot(x='Day', y='rmean', data=rolling_stats, label='Rolling Average')
sns.lineplot(x='Day', y='rstd', data=rolling_stats, label='Rolling Standard Deviation')
```

## Analysis of Time Series 

The following report of findings was performed on the entire time series and not just the training set. Please see the next few pages for results for autocorrelation, seasonality, trend, spectral density, decomposed time series, trends in residuals of decomposed series, and differencing. 

#### Autocorrelation

Autocorrelation is the correlation between the time series and a lagged version of the same series. Plotting autocorrelation shows that the time series is not stationary since every lag is outside of the confidence interval. In all plots the first lags are larger than subsequent lags which indicates a trend (Hyndman et al, 2018). 

#### Partial Auto Correlation

Partial autocorrelation is the correlation between a point and the same point in the lagged time series. The partial auto correlation plot below can help determine the autoregressive term p in the ARIMA model by counting lags that are far outside the confidence interval (Analytics India, 2024). 

```python
# Plot ACF with differencing (Analytics India, 2024).
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(revenue, alpha=.05)
print()
```

```python
# Plot Partial Auto Correlation
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(revenue)
print()
```

#### Spectral Density 

The spectral density is a map of the covariance of time and revenue in the time series and how they vary together. It can show underlying periodicity or seasonality in the data. In the chart below of the power spectral density of the time series. There seems to be one large spike which indicates these variables vary together at a smaller frequency. 

```python
# Plot the spectral density 

plt.psd(revenue['Revenue'])
plt.title('Spectral Density of Time Series')
```

#### Seasonal Decomposition 

Seasonal decomposition breaks the time series down into the the underlying elements of the series such as Trend, Seasonality and Residuals. The main plot below shows all elements together. Each are broken down and discussed individually after. 

```python
# import libraries for seasonal decomposition 
from statsmodels.tsa.seasonal import seasonal_decompose

# perform decomposition on the time series
decomposed = seasonal_decompose(revenue['Revenue'], model = 'additive')
```

```python
# Plot each of the decomposed elements 
decomposed.plot()
print()
```

However, the direction of the trend is not apparent from the ACF. Looking at the decompositon performed above we can look at the trend plot more closely to identify the direction of the trend. The revenue trend is overall increasing in the two year window of observations of the revenue.

#### Differencing

Since this time series is not stationary, the ARIMA model will require a parameter for differencing to make the time series stationary. Using differencing can help determine what degree of differencing will be necessary to make the time series stationary for the analysis. First order differencing on the time series is plotted below. The Dickey Fuller test for stationarity shows a p value for the differenced series of 0.0 which means the first order differencing has made the time series stationary. The trend is no longer visible in the data. 

```python
# Difference Training set to obtain stationarity
diff = revenue.diff().dropna()
plt.figure(figsize = [12,6])
diff.plot()
```

```python
# Dickey Fuller Test for Stationarity 
dftest = adfuller(diff, autolag='AIC', regression='ct')
print("The P-Value for the DF test is: ", dftest[1])
if dftest[1] < .05:
    print("The p value is less than .05, reject null hypothesis. Differenced series is stationary.")
else :
    print("The p value is greater than .05, accept null hypothesis. Differenced series is non-stationary.")
```

### Development of ARIMA Model Including Output, Code, and Forecast 

I used the following process to identify the ARIMA model that takes into account seasonality and trend: 

The seasonality identified in the time series was weekly, the period is expected to repeat every 7 days. Using auto arima with seasonal settings on , I identified the parameters that produced a SARIMAX model with the best performing AIC. Fitting the model on the training set I plotted the predicted values against the actual training set and then forecast 90 days from the training set and compared it against the actual values of the test set. At this point I re-evaluated the model considering what was learned in earlier analysis of the time series regarding seasonality and trend. I adjusted the parameters to produce a better AIC and more accurate forecast. This final model was used for the remainder of the analysis. The final model was fit to the test set and predictions plotted against actual values. Finally, the model was fitted to the entire dataset to predict the next 30 days of revenue.

```python
import pmdarima as pmd 

# Using arima to find SARIMAX parameters for p, d, q, P, D, & Q
seasonal_model = pmd.auto_arima(revenue['Revenue'], start_p = 0, start_q = 0, test='adf', trace=True, seasonal=True, m=7)
```

```python
import statsmodels.api as sm

# Fit SARIMAX model to the training set using parameters from auto_arima
smodel=sm.tsa.statespace.SARIMAX(X_train,order=(1, 0, 0),seasonal_order=(1,0,1,7))
sresults=smodel.fit(disp=0)

# View summary of results of fit
sresults.summary()
```

```python
# Plot predicted values against actual training values

sprediction = sresults.predict()
plt.title('Predicted Values vs. Actual Training Values')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.plot(sprediction)
plt.plot(X_train)
```

```python
# Plot diagnostics for model 

sresults.plot_diagnostics()
print()
```

```python
#Obtain forecasted values for 90 days from training set to compare against actual test set

sforecast = sresults.get_forecast(steps =90, dynamic=True)
smean_forecast = sforecast.predicted_mean
sconf_intervals = sforecast.conf_int()

# Plot forecast against test and training set with confidence intervals
plt.figure(figsize = [12 , 6])
plt.title('Initial Model: Mean 90 Forecast vs. Actual Test Values')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.plot(X_train['Revenue'], label = 'Train Set')
plt.plot(X_test['Revenue'], label = 'Test Set')
plt.plot(smean_forecast)
plt.fill_between(sconf_intervals.index, sconf_intervals.iloc[:, 0],
                sconf_intervals.iloc[:, 1], color='k', alpha=.2)
```

#### Before Continuing Analysis, Addressing Problems with Model 

This doesn't look very good. The confidence intervals do capture the test set, but the all over mean forecast is down trending when the true data trends up. This is very likely because the model selected by auto_arima includes no differencing. Since the time series is not stationary, a differencing factor should be included. In the notes above, a first order differencing is sufficient to make the time series stationary. Yet since differencing the dataset removes trend, we should also include a seasonal differencing to offset this problem. 


The AIC of the model below that includes differencing is much improved. In the diagnostic histogram of residuals you can see that the residuals are more closely centered around 0. But most importantly, the forecast captures an upward trend in the test data more accurately. This will be used as my final model. The rest of the analysis will follow including running the final model on the test set and projecting revenue beyond the total time series. 


```python
# SARIMAX Model that includes differencing 

final_model=sm.tsa.statespace.SARIMAX(X_train,order=(1, 1, 0),seasonal_order=(1,1,1,7))
final_results=final_model.fit(disp=0)
final_results.summary()
```

```python
# Plot predicted values against the training set 

final_prediction = final_results.predict()
plt.title('Predicted Values vs Actual Training Values')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.plot(final_prediction, label = 'Prediction')
plt.plot(X_train, label = 'Test Data')
plt.legend()
```

```python
# Print diagnostic plots for final model
final_results.plot_diagnostics()
print()
```

```python
# Forecast 90 days from training set to compare against test set
final_forecast = final_results.get_forecast(steps =90, dynamic=True)
final_mean_forecast = final_forecast.predicted_mean
final_conf_intervals = final_forecast.conf_int()

# Plot forecast against test and training set
plt.figure(figsize = [12 , 6])
plt.title('Final Model: Mean 90 Forecast vs. Actual Test Values')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.plot(X_train['Revenue'], label = 'Train Set')
plt.plot(X_test['Revenue'], label = 'Test Set')
plt.plot(final_mean_forecast, label = 'Forecast')
plt.fill_between(final_conf_intervals.index, final_conf_intervals.iloc[:, 0],
                final_conf_intervals.iloc[:, 1], color='k', alpha=.2)
plt.legend()
```

#### Mean Square Error and Other Evaluation Metrics 

The mean squared error which square the difference between the actual values and the predicted values is 16.82. The mean squared error often over emphasizes the larger errors due to squaring so the mean absolute error and root mean squared error were also calculated. The mean absolute error is an average of the absolute difference between the predicted values and the actual values. It is less sensitive than the mean squared error to outliers. The mean absolute error for the forecasted values compared to the actual test values is 3.46. The square root of the mean squared error is 4.10. Both of these last metrics are expressed in the same units as the revenue variable and are more easily interpreted than the mean squared error. 


```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math

# Print Mean Squared Error, Absolute Mean Error, and Root Mean Squared Error (Dhaduk, H.).
mse = mean_squared_error(X_test, final_mean_forecast)
print('MSE: '+str(mse))
mae = mean_absolute_error(X_test, final_mean_forecast)
print('MAE: '+str(mae))
rmse = math.sqrt(mean_squared_error(X_test, final_mean_forecast))
print('RMSE: '+str(rmse))

```

#### Fitting Model to Full Time Series to Predict 30 Days Revenue 

In the following steps the model is fitted to the entire time series to predict 30 days past the original time series. The forecast is calculated and plotted with confidence intervals. 

```python
# Fit model to the entire dataset
pmodel=sm.tsa.statespace.SARIMAX(revenue,order=(1, 1, 0),seasonal_order=(0,1,0,7))
presults=pmodel.fit(disp=0)
presults.summary()
```

```python
# Forecast data points for 30 days out from the entire time series 
revenue_forecast = presults.get_forecast(steps =30, dynamic=True)
revenue_mean_forecast = revenue_forecast.predicted_mean
revenue_conf_intervals = revenue_forecast.conf_int()

# Plot forecast against time series, highlighting the test portion and confidence intervals. 
plt.figure(figsize = [12 , 6])
plt.title('30 Day Final Revenue Forecast')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.plot(revenue['Revenue'], label = 'Train Set')
plt.plot(X_test['Revenue'], label = 'Test Set')
plt.plot(revenue_mean_forecast, label = '30 Day Forecast')
plt.fill_between(revenue_conf_intervals.index, revenue_conf_intervals.iloc[:, 0],
                revenue_conf_intervals.iloc[:, 1], color='k', alpha=.2)
plt.legend()
```

### Justification of Forecast Length

ARIMA models are not reliable for long-term forecasts and have difficulty predicting future turning points. This is something I noticed in my own experimenting with forecast lengths during this project. Once the model has decided on the slope of the forecast there doesn't seem to be much to change its mind. ARIMA forecasts are founded entirely on past performance which can't always predict future performance. 

<u>For the forecast taken from the training set predicted on the interval of the training set</u>

I deliberately chose the length of the forecast to match the length of the test set so that the two could be compared. This was done specifically to validate the predictive power of the model fitted to the training set to compare to actual values. It made visual comparison of the forecast against actual values much more easy to interpet. 

<u>For the forecast taken from the whole data set predicted for a 30 day interval in the future</u>

For this forecast I wanted to make the metric as small and modest as possible to take into account the limitations of ARIMA models. But I also wanted to make it long enough that it was useful from a business perspective. A month prediction can be useful in terms of budgeting and making short term business plans.
